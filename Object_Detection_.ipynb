```python
# Step 1: Install Ultralytics (YOLOv8)
!pip install ultralytics --quiet

# Step 2: Import Required Libraries
from ultralytics import YOLO
import matplotlib.pyplot as plt
from google.colab import files
from PIL import Image
import os
import glob

# Step 3: Load Pretrained YOLOv8n Model
model = YOLO("yolov8n.pt")  # lightweight version

# Step 4: Upload Image (jpg, jpeg, png)
print("ğŸ“¤ Upload an image (jpg, jpeg, or png):")
uploaded = files.upload()

# Step 5: Predict and Display
for filename in uploaded.keys():
    ext = filename.split('.')[-1].lower()
    if ext not in ['jpg', 'jpeg', 'png']:
        print(f"âŒ Unsupported file format: {filename}")
        continue

    # Step 6: Run Prediction
    results = model.predict(source=filename, conf=0.3, save=True)

    # Step 7: Locate the saved image path from latest run
    save_dir = results[0].save_dir  # automatically gets latest /predict, /predict2, etc.
    detected_images = glob.glob(f"{save_dir}/*")

    if detected_images:
        detected_img_path = detected_images[0]
        print("âœ… Detection Complete! Displaying result...")
        image = Image.open(detected_img_path)
        plt.figure(figsize=(10, 10))
        plt.imshow(image)
        plt.axis("off")
        plt.title("ğŸ” YOLOv8 Object Detection Result")
        plt.show()
    else:
        print("âš ï¸ No detection image found.")
```
